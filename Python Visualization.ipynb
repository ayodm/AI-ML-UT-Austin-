```python
import pandas as pd
import seaborn as sns  # sns = reference to the west wing
import matplotlib.pyplot as plt # Seaborn is based on matplotlib
sns.set(color_codes=True) # Add clean background to graphs
%matplotlib inline

```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

## Plotting univariate distributions

The most convenient way to take a quick look at a univariate distribution in seaborn is the distplot() function. By default, this will draw a histogram and fit a kernel density estimate (KDE)

```python
sns.displot(auto['highway_mpg']);
```

```python
sns.displot(auto['highway_mpg'], kde=False, rug=True);
```

```python
# We can turn the KDE off and put a tic mark along the x-axis for every data point with rug
sns.displot(auto['city_mpg'],kde=False, rug=True);
```

## Plotting bivariate distributions 

It can also be useful to visualize a relationship between two variables. The easiest  way to do this in seaborn is to use the jointplot() function which is a scatterplot of the two variables along with the histograms of each next to the appropriate axes. 

```python
sns.jointplot(auto['engine_size'], auto['horsepower']);
```

## Hex Bin Plots
 We can make a hex bin plot that breaks the 2D area into hexagons and the number of points in each hexagon determines the color
 

```python
x = auto['engine_size']
y = auto['horsepower']
sns.jointplot(x,y,kind="hex");
```

## Kernel Density Estimation 
We can make a 2D estimation of density also/

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="kde");
```

## Visualizing pairwise relationships in a dataset

To plot multiple pairwise scatterplots in a dataset, you can use the pairplot() function. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame, it also draws the histogram of each variable on the diagonal Axes: 

Be careful about toggle scrolling in the cell menu

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']]);
```

## Plotting with categorical data

In the strip plot, the scatterplot points will usually overlap. This makes it difficult to see the full distribution of data. One easy solution is to adjust the distribution( along the categorical axis) using 'jitter'

```python
sns.stripplot(auto['fuel_type'],auto['horsepower'], jitter=True);
# jitter spreads out the data a little bit so we can have a better representation of how many data points there are
```

A different Approach would be to use the function swarmplot(), which postiions each scatterplot point on the categorical axis and avoids overlapping. 

```python
sns.swarmplot(auto['fuel_type'],auto['horsepower']);
# A swarmbot uses some math to figure out how far we need to spread the scatter data points 
```

## Boxplots 

Another common graph, similar to a histogram but instead of bars for every single bin just a few points are noted. 
This kind of plot shows the three quartile values of the distribution along with exterme values. The "whiskers" extend to points that lie within 1.5 IQRs of the lower upper quartile, and then observations that fall outside this range are displayed independently. 


```python
sns.boxplot(auto['number_of_doors'],auto['horsepower']);
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type']);
```

## Bar Plots 

We can plot the mean of a dataset, separated in categories using the barplot() function. When there are multiple observations in each category, it uses bootstrapping to compute a confidence interval around the estimate and plots that using error bars: 

Bar plots start at 0, which can sometimes be practival if zero is a number you want to compare to

```python
sns.barplot(auto['body_style'],auto['horsepower'], hue=auto['fuel_type']);
```

```python
sns.countplot(auto['body_style'], hue=auto['fuel_type']);
```

## Point Plots

An alternative style for visualizing the same information is offered by the pointplot() functions. This function also encodes the value of the estimate with height on the other axis, but rather than show a full bar it just plots the point estimate and confidence interval. Additionally, pointplot connects points from the same hue category. This makes it easy to see how the manin relationship is changing as a function of a second variable, because your eyes are quite good at picking up on differences of slopes:

```python
x1 = auto['body_style']
y1 = auto['horsepower']
sns.pointplot(x1, y1, hue=auto['number_of_doors']);

# sns.pointplot(auto['body_style'],auto['horsepower'],hue=auto['number_of_doors']);
```

```python
help(sns.pointplot())
```

## Drawing multi-panel categorical plots 


```python
sns.catplot(x="fuel_type",
           y="horsepower",
           hue="number_of_doors",
           col="drive_wheels",
            data=auto,
           kind="box");
```

```python
sns.catplot(x="fuel_type",
           y="horsepower",
           hue="number_of_doors",
           col="drive_wheels",
            data=auto,
           kind="violin");
```

Various type of kind input: {point, bar, count, box, violin, strip}

## Function to draw linear regression models 

lmplo() is one of the most widely used functions to quickly plot the Linear Relationship between 2 variables

```python
sns.lmplot(y="horsepower", x="engine_size", data=auto);
```

```python
sns.lmplot(y="horsepower", x="engine_size", hue="fuel_type", data=auto);
```
