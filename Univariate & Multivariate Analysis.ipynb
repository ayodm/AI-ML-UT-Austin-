# Univariate Analysis

Refers to the analysis of a single variable. THe main purpose of univariate analysis is to summarize and find patterns in the data. The key point is that htere is only one variable involved in the analysis. 


```python
# The code below plots a histogram using the matplotlib package. 
# the bins argument create class intervals. In this case we are creating 50 bins. 

plt.hist(data, bins=50)
```

In the Above histogram, the first array is the frequency in each class and the second array contains the edges of the class intervals. These arrays can be assigned to a variable and used for further analysis. 

```python
sns.displot(data=data, kde=True)
```

```python
sns.distplot(data, hist=False) # plots only the frequency
```

```python
sns.violinplot(data) 
# gives us a focus on the # of houses, based on distance
```

Now let us have a closer look at the distribution by plotting a simple histogram with 10 bins

```python
plt.figure(figsize=(20,10)) # makes the plot wider
plt.hist(data, color='g') # plots a simple histogram
plt.axvline(data.mean(), color='r', linewidth=2)
plt.axvline(data.median(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(data.mode()[0], color='w', linestyle='dashed',linewidth=1)
```

* In the histogram above we can see that the second bin with class limit(4.81, 9.62) has the most number of observations. 
* A bulk of observations lie within the first four classes. the rest of the six classes contain fewer observations. 
* Since we are measuring distance. We can assume based on our visual observation that majority of the data points are less than 20kms
* There are very few distances greater than 20 kms, past 30km even fewer
* The mean is represented by the red line, median by the blue line, mode by the white line
* The mean and the mode are very close to each other and the Median is less than them 

**Now we have an idea of how the data is distributed**

# Multivariate Analysis

Performed to understand the interactions between different fields in the dataset(or) finding the interactions between variables more than 2

ex: Pairplot, 3D scatter plot

# Scatter Plot

Gives us an idea of the association between two variables


```python
data1 = pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
data1.head()
```

```python
data1 = data1.dropna()
```

```python
sns.pairplot(data1)
```

**In the Above plot scatter diagrams are plotted for all the numerical columns in the dataset. A scatter plot is a visual representation of the degree of correlation between any two columns. The pair plot function in seaborn makes it very easy to generate joint scatter plots for all the columns in the data.**

**A scatter plot can also be plotted for two individual columns:**

```python
sns.scatterplot(data1['Price'],data1['Distance']) 
```

The scatter plot only offers visual information about the degree of correlation. In order to obtain more precise information we can us the inbuilt .corr() method in Pandas. This returns a table with all the correlations calculated for the numerical columns 

```python
data1.corr() # displays the correlation between every possible pair of attributes as a dataframe
```

```python
sns.heatmap(data1.corr(), annot=True) 
```
