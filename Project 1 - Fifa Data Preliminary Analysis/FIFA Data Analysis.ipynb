* Change the datatype for any columns in FIFA dataset that are 'object' to 'category'

### Fix Datatype(s)

We want to change any data types that are objects to category because the data is categorical. This also helps our dataframe manager storage of data better.

```python
fifa[fifa.
     select_dtypes(['object']).
     columns] = fifa.select_dtypes(['object']).apply(lambda x: x.astype('category'))
```
