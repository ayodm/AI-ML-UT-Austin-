## Interpret a particular age

```python

Players_D = fdata[["Age", "Overall", "Potential"]].copy()
sortedbyage = Players_D.sort_values(by="Age")
sortedbyage.info()

Players_D.to_csv("Players_D.csv", index=True)
```

```python
sortedbyage.describe(include="all")
```

```python
groupedbyage = Players_D.groupby('Age')
print(groupedbyage)

groupedbyage.describe()
```

```python
sns.distplot(Players_D['Age'], kde=False, rug=True);
```

```python
sns.jointplot(Players_D['Age'],Players_D['Overall']);
sns.jointplot(data = Players_D, y='Age',x='Overall', kind='reg');

g = sns.JointGrid(data=Players_D, x='Age', y='Overall')
g.plot(sns.regplot, sns.boxplot)
```
