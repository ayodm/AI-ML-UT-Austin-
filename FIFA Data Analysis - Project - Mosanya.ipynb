## Analysis of offensive players

```python
OPlayers = fdata[['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
    'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM',
    'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB',
    'LCB', 'CB', 'RCB', 'RB', 'Wage']].copy()


def position_to_num(pos_val):
    """For each value, take the number before the '+'
    unless it is not a string value. This will only happen
    for NaNs so in that case we just return NaN.
    """
    
    if isinstance(pos_val,str):
        return float(pos_val.split('+')[0])
    else: 
        return np.nan
    
position_cols = [
    'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
    'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM',
    'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB',
    'LCB', 'CB', 'RCB', 'RB'
]

for colname in position_cols:
    OPlayers[colname] = df[colname].apply(position_to_num)
    
    
# Replacing Position Names 
    
off_pos_to_cols = {
    'Striker': ['ST', 'CF','LS','RF','LF', 'RS'], 
    'Right-Winger': ['RW'],
    'Left-Winger': ['LW']
#    'Left-Midfield':['LM', 'LAM'],
#    'Right-Midfield':['RM', 'RAM'],
#    'Right-DefMid':['RDM'],
#    'Left-DefMid':['LDM'],
#    'Center-Mid':['CM','CAM'],
#    'Center-DefMid':['CDM'],
#    'Right-Defense':['RB','RCB'],
#    'Center-Def':['CB'], 
#    'Left-Defense':['LB','LCB']    
}

for pos_type, colvec in off_pos_to_cols.items():
    OPlayers[pos_type +'_rate']= round(df[colvec].mean(axis=1))
    
    
OPlayers.drop(position_cols, axis=1, inplace=True)

print(OPlayers.shape)
print(OPlayers)
```
