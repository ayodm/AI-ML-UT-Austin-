## Analysis of clubs with maximum share of players from England, Spain & Germany (Top 10)

```python
fdata.head()
data = fdata[["Club", "Nationality", "Name"]]

filter3 = data["Nationality"] == "England"
England = data.where(filter3).dropna()
#print(England)
print("Country: England")
print("\n")
print(England['Club'].value_counts().nlargest(10))
print("\n")
print("\n")

filter2 = data["Nationality"] == "Spain"
Spain = data.where(filter2).dropna()
#print(Spain)
print("Country: Spain")
print("\n")
print(Spain['Club'].value_counts().nlargest(10))
print("\n")
print("\n")

filter1 = data["Nationality"]=="Germany"
Germany = data.where(filter1).dropna()
#print(Germany)
print("Country: Germany")
print("\n")
print(Germany['Club'].value_counts().nlargest(10))

```
